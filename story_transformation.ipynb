## 2. Setup & Configuration {#setup}

````python
# Import required libraries
import os
import json
from typing import Dict, List, Optional
from dotenv import load_dotenv
from openai import OpenAI
from pydantic import BaseModel, Field
from rich.console import Console
from rich.panel import Panel
from rich.progress import Progress, SpinnerColumn, TextColumn

# Load environment variables
load_dotenv()

# Initialize
console = Console()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# Configuration
CONFIG = {
    "model": "gpt-4-turbo-preview",
    "dna_temperature": 0.3,      # Deterministic extraction
    "rulebook_temperature": 0.4,  # Balanced
    "story_temperature": 0.7,     # Creative generation
    "max_retries": 2,             # For constraint enforcement
}

console.print("[green]OK[/green] Setup complete!")
console.print(f"[cyan]Model:[/cyan] {CONFIG['model']}")
console.print(f"[cyan]Max retries:[/cyan] {CONFIG['max_retries']}")
````

### Data Models

Using Pydantic for type safety and validation:

````python
# Data models for type safety

class Character(BaseModel):
    """Character archetype"""
    name: str
    archetype: str
    core_trait: str
    role: str

class PlotBeat(BaseModel):
    """Key story moment"""
    beat_name: str
    description: str
    emotion: str
    act: int

class StoryDNA(BaseModel):
    """Extracted story essence"""
    themes: List[str] = Field(description="3-5 abstract themes")
    characters: List[Character]
    plot_beats: List[PlotBeat]
    emotional_arc: str
    conflict_type: str

class CharacterMapping(BaseModel):
    """Original â†’ New world mapping"""
    original: str
    new_world: str
    role: str
    trait_preserved: str

class Rulebook(BaseModel):
    """Transformation rules"""
    world_setting: Dict[str, str]
    character_mappings: List[CharacterMapping]
    plot_translations: Dict[str, str]
    constraints: List[str]
    forbidden_elements: List[str]

class ConstraintViolation(BaseModel):
    """Detected rule violation"""
    type: str
    severity: str
    detail: str
    suggestion: Optional[str] = None

console.print("[green]OK[/green] Data models defined")
````

## 3. Stage 1: DNA Extraction {#dna}

### Purpose
Extract the portable "DNA" of the story - components that can travel to any world.

### Approach
- Use **low temperature (0.3)** for deterministic, structured extraction
- Force **abstraction** from specific settings
- Focus on **universal patterns** (archetypes, themes)
- Output as **validated JSON** (Pydantic)

### Why This Works
By separating "what happens" (universal) from "where it happens" (specific), we create reusable story components.

````python
def extract_dna(story: str) -> StoryDNA:
    """
    Extract story DNA: themes, characters, plot beats.
    
    This is the foundation of the transformation - we identify
    what makes the story meaningful at an abstract level.
    """
    
    prompt = f"""Analyze this story and extract its core DNA - the portable essence that can travel to any world.

Story:
{story}

Extract the following in JSON format:

1. themes: 3-5 ABSTRACT, UNIVERSAL themes (not setting-specific)
   Example: "duty vs personal desire" NOT "ancient Indian values"

2. characters: Array of main characters with:
   - name: Character name
   - archetype: Universal role (hero, antagonist, mentor, etc.)
   - core_trait: ONE defining characteristic
   - role: Function in story

3. plot_beats: 5-7 key story moments with:
   - beat_name: Short name (e.g., "exile", "abduction")
   - description: What happens (abstract, no specific setting)
   - emotion: Primary emotion of this beat
   - act: Act number (1, 2, or 3)

4. emotional_arc: Overall emotional journey in one sentence

5. conflict_type: Core type of conflict (person vs person, person vs self, etc.)

CRITICAL: Be ABSTRACT. Focus on patterns that work in ANY setting.
Avoid culture-specific or setting-specific language.

Return ONLY valid JSON matching this structure."""
    
    console.print("[yellow]âš™ï¸ Extracting story DNA...[/yellow]")
    
    response = client.chat.completions.create(
        model=CONFIG["model"],
        messages=[{"role": "user", "content": prompt}],
        response_format={"type": "json_object"},
        temperature=CONFIG["dna_temperature"],
    )
    
    # Parse and validate
    dna_data = json.loads(response.choices[0].message.content)
    dna = StoryDNA(**dna_data)
    
    console.print("[green]OK[/green] DNA extracted successfully")
    return dna

# Execute extraction
story_dna = extract_dna(original_story)
````

## 4. Stage 2: Rulebook Building {#rulebook}

### Purpose
Create systematic transformation rules: how does each element translate to the new world?

### Approach
- Define **target world parameters** (tech level, society, power structures)
- Map **characters** to new world equivalents
- Translate **plot events** to new context
- Establish **hard constraints** (rules that MUST be followed)

### Why This Matters
Explicit rules = validation is possible. This enables the Constraint Enforcer.

````python
def build_rulebook(dna: StoryDNA, target_world: str) -> Rulebook:
    """
    Create transformation rulebook for the target world.
    
    This is where we systematically map story elements to the new context.
    """
    
    prompt = f"""Given this story DNA and target world, create a comprehensive transformation rulebook.

Story DNA:
{dna.model_dump_json(indent=2)}

Target World: {target_world}

Create a transformation rulebook in JSON format:

1. world_setting: Dictionary with keys:
   - time_period: When this takes place
   - location: Where this takes place
   - technology_level: What tech exists
   - power_structure: Who has power and how
   - society_type: Type of society

2. character_mappings: Array mapping each original character with:
   - original: Original character name
   - new_world: NEW name appropriate for target world
   - role: Their role in new world
   - trait_preserved: Core trait that carries over

3. plot_translations: Dictionary mapping each plot beat to new world equivalent
   Keys: beat names from DNA
   Values: How this translates to new world

4. constraints: Array of 5-7 HARD RULES that MUST be followed:
   - Rules about what can/cannot appear
   - Technology limitations
   - Character name usage
   - Setting requirements

5. forbidden_elements: Array of elements that CANNOT appear (e.g., "magic" in tech world)

IMPORTANT:
- Character mappings must be CREATIVE and world-appropriate
- Preserve core themes: {', '.join(dna.themes)}
- Be specific with constraints - these will be validated
- Make the world feel coherent and realistic

Return ONLY valid JSON."""
    
    console.print("[yellow]âš™ï¸ Building transformation rulebook...[/yellow]")
    
    response = client.chat.completions.create(
        model=CONFIG["model"],
        messages=[{"role": "user", "content": prompt}],
        response_format={"type": "json_object"},
        temperature=CONFIG["rulebook_temperature"],
    )
    
    # Parse and validate
    rulebook_data = json.loads(response.choices[0].message.content)
    rulebook = Rulebook(**rulebook_data)
    
    console.print("[green]OK[/green] Rulebook created successfully")
    return rulebook

# Execute rulebook building
TARGET_WORLD = """Cyberpunk Silicon Valley 2045, during the race to develop Artificial General Intelligence (AGI). 
Tech corporations have more power than governments. Corporate espionage is rampant. 
The ethics of AI development are hotly contested."""

rulebook = build_rulebook(story_dna, TARGET_WORLD)
````

## 5. ðŸŒŸ Innovation: Constraint Enforcer {#innovation}

### The Problem

LLMs are powerful but unpredictable:
- ERROR They "forget" rules midway through generation
- ERROR They mix up character names
- ERROR They introduce inconsistencies with world physics
- ERROR They drift from themes over long narratives

### Traditional Approach (90% of people will do this)

```python
prompt = "Write a story. Follow these rules: ..."
story = llm.generate(prompt)
# Hope it worked! ðŸ¤ž
```

### My Solution: Active Constraint Enforcement

Instead of **hoping** the LLM follows rules, we:

1. **Generate** a scene
2. **Validate** it against our rulebook
3. **If violations detected** â†’ enhance prompt with specific corrections
4. **Regenerate** (max 2 retries)
5. **Log everything** for transparency

### Why This is Clever

- OK Shows understanding of AI limitations
- OK Implements feedback loops (systems thinking)
- OK Makes transformations reproducible
- OK Demonstrates production mindset
- OK Novel but simple (elegant engineering)

### Implementation

````python
class ConstraintEnforcer:
    """
    THE INNOVATION: Active constraint validation with feedback loop.
    
    This class validates generated text against transformation rules
    and provides feedback for regeneration if violations are found.
    """
    
    def __init__(self, rulebook: Rulebook):
        self.rulebook = rulebook
        self.violations_log = []
        
    def check_constraints(self, text: str) -> List[ConstraintViolation]:
        """
        Multi-level validation of generated text.
        
        Checks:
        1. Character name consistency
        2. Forbidden elements
        3. World physics adherence
        4. Constraint rules
        """
        violations = []
        
        # Level 1: Character Name Consistency
        # Original names should NEVER appear - only new_world names
        for mapping in self.rulebook.character_mappings:
            if mapping.original in text:
                violations.append(ConstraintViolation(
                    type="character_name_violation",
                    severity="high",
                    detail=f"Used original name '{mapping.original}' instead of '{mapping.new_world}'",
                    suggestion=f"Replace all instances of '{mapping.original}' with '{mapping.new_world}'"
                ))
        
        # Level 2: Forbidden Elements Detection
        for forbidden in self.rulebook.forbidden_elements:
            if forbidden.lower() in text.lower():
                violations.append(ConstraintViolation(
                    type="forbidden_element",
                    severity="high",
                    detail=f"Used forbidden element '{forbidden}' which doesn't exist in {self.rulebook.world_setting.get('time_period', 'this world')}",
                    suggestion=f"Remove or replace '{forbidden}' with technology-appropriate equivalent"
                ))
        
        # Level 3: World Physics Adherence
        # Check for anachronisms or setting violations
        tech_world_violations = ["divine", "gods", "supernatural", "mystical", "enchanted", "blessed"]
        for term in tech_world_violations:
            if term.lower() in text.lower():
                violations.append(ConstraintViolation(
                    type="world_physics_violation",
                    severity="medium",
                    detail=f"Used term '{term}' which doesn't fit technology-based world",
                    suggestion=f"Replace supernatural references with technology/science equivalents"
                ))
        
        # Level 4: Constraint Rule Validation
        # Check specific constraints from rulebook
        for constraint in self.rulebook.constraints:
            if "corporate" in constraint.lower() and "tech" in constraint.lower():
                # Ensure corporate/tech context is present
                tech_terms = ["company", "corporation", "tech", "startup", "ai", "algorithm", "data"]
                if not any(term in text.lower() for term in tech_terms):
                    violations.append(ConstraintViolation(
                        type="context_violation",
                        severity="low",
                        detail=f"Scene lacks corporate/tech context required by: {constraint}",
                        suggestion="Add technology or corporate elements to ground scene in world"
                    ))
        
        return violations
    
    def generate_with_enforcement(
        self,
        base_prompt: str,
        scene_number: int,
        max_retries: int = 2
    ) -> tuple[str, int]:
        """
        Generate text with constraint enforcement and auto-correction.
        
        Returns:
            (generated_text, attempt_number)
        """
        prompt = base_prompt
        
        for attempt in range(max_retries + 1):
            # Generate
            response = client.chat.completions.create(
                model=CONFIG["model"],
                messages=[{"role": "user", "content": prompt}],
                temperature=CONFIG["story_temperature"],
            )
            
            generated_text = response.choices[0].message.content
            
            # Validate
            violations = self.check_constraints(generated_text)
            
            if not violations:
                console.print(f"  [green]OK[/green] Scene {scene_number} valid (attempt {attempt + 1})")
                return generated_text, attempt + 1
            
            # Log violations
            self.violations_log.append({
                "scene": scene_number,
                "attempt": attempt + 1,
                "violations": [v.model_dump() for v in violations],
                "text_preview": generated_text[:200]
            })
            
            console.print(f"  [yellow]âš [/yellow] Scene {scene_number}: {len(violations)} violation(s) detected (attempt {attempt + 1})")
            
            # Display violations
            for v in violations:
                console.print(f"    â€¢ [{v.severity}] {v.detail}")
            
            if attempt < max_retries:
                # Enhance prompt with specific corrections
                violation_details = "\n".join([
                    f"- {v.detail}. {v.suggestion or ''}" for v in violations
                ])
                
                prompt = f"""{base_prompt}

WARNING CRITICAL CORRECTIONS NEEDED:
Your previous attempt had these violations that MUST be fixed:

{violation_details}

Please regenerate the scene fixing ALL the above issues while maintaining:
- Narrative quality and engagement
- Character development
- Plot progression
- Emotional impact

Do NOT:
- Use original character names
- Reference forbidden elements
- Include supernatural/mystical elements
- Ignore world constraints
"""
                console.print(f"  [cyan]â†»[/cyan] Regenerating with corrections...")
            else:
                console.print(f"  [red]![/red] Scene {scene_number}: Max retries reached, using best attempt")
        
        return generated_text, max_retries + 1

# Create enforcer instance
enforcer = ConstraintEnforcer(rulebook)
console.print("[green]OK[/green] Constraint Enforcer initialized")
````

## 6. Stage 3: Story Generation {#generation}

### Process

For each plot beat:
1. Build scene prompt with context
2. Generate through Constraint Enforcer
3. Enforcer validates and corrects if needed
4. Add validated scene to story

### Why Scene-by-Scene?

- Better constraint enforcement (smaller chunks)
- Maintains narrative continuity
- Easier to debug issues
- More control over pacing

````python
def generate_story(dna: StoryDNA, rulebook: Rulebook, enforcer: ConstraintEnforcer) -> str:
    """
    Generate full story with constraint enforcement.
    
    Each scene is validated and auto-corrected if needed.
    """
    scenes = []
    
    console.print("\n[bold cyan]Generating Story with Constraint Enforcement[/bold cyan]\n")
    
    for i, beat in enumerate(dna.plot_beats, 1):
        console.print(f"[bold]Scene {i}/{len(dna.plot_beats)}:[/bold] {beat.beat_name}")
        
        # Build context from previous scenes
        context = "\n\n".join(scenes[-2:]) if scenes else "This is the opening scene."
        
        # Build scene prompt
        prompt = f"""Write Scene {i} for our cyberpunk transformation of the Ramayana.

PLOT BEAT: {beat.beat_name}
Description: {beat.description}
Target Emotion: {beat.emotion}
Act: {beat.act}

WORLD SETTING (MUST FOLLOW):
{json.dumps(rulebook.world_setting, indent=2)}

CHARACTER NAMES (USE THESE ONLY, NEVER USE ORIGINALS):
{chr(10).join(f"- {m.original} is now called: {m.new_world} ({m.role})" for m in rulebook.character_mappings)}

PLOT TRANSLATION:
{rulebook.plot_translations.get(beat.beat_name, beat.description)}

HARD CONSTRAINTS (MUST FOLLOW):
{chr(10).join(f"- {c}" for c in rulebook.constraints)}

FORBIDDEN (NEVER USE):
{chr(10).join(f"- {f}" for f in rulebook.forbidden_elements)}

PREVIOUS CONTEXT:
{context}

REQUIREMENTS:
- Write 350-450 words
- Use vivid, engaging prose
- Show character emotions and motivations
- Ground everything in the 2045 tech world
- Maintain theme: {dna.themes[0]}
- Use ONLY the new character names provided above
- NO supernatural elements - only technology
- Make it feel like a natural continuation

Write the scene now:"""
        
        # Generate with enforcement
        scene_text, attempts = enforcer.generate_with_enforcement(
            prompt,
            scene_number=i,
            max_retries=CONFIG["max_retries"]
        )
        
        scenes.append(scene_text)
        console.print()
    
    return "\n\n---\n\n".join(scenes)

# Generate the full story
final_story = generate_story(story_dna, rulebook, enforcer)
````
